```python
w=np.zeros(1)
b=0
```

```python
def cost_function(x,y,w,b):
    f_wb=np.dot(x,w)+b
    cost= np.sum((f_wb-y)**2)/(2*x.shape[0])
    return cost
```

```python
print(cost_function(x_train,y_train,w,b))
```

```python
def calculate_derivative(x,y,w,b):
    error=(np.dot(x,w)+b)-y
    dj_db=np.sum(error)/(x.shape[0])
    dj_dw=np.dot(error,x)/(x.shape[0])
    return dj_dw,dj_db
```

```python
dj_dw, dj_db=calculate_derivative(x_train,y_train,w,b)
print(dj_dw,dj_db)
```

```python
def gradient_descent(x,y,w,b,a,it):
    for i in range(it):
        dj_dw,dj_db=calculate_derivative(x,y,w,b)
        w-= a*dj_dw
        b-= a*dj_db
    return w,b
```

```python
gradient_descent(x_train,y_train,w,b,1e-7,10000)
```

```python
def prediction(test,x,y,a,it):
    m,n=x.shape
    w=np.zeros(n)
    b=0
    w,b=gradient_descent(x,y,w,b,a,it)
    f_wb= np.dot(test,w)+b
    return f_wb
```

```python
pred=prediction(x_train,x_train,y_train,0.5e-7,500000)
```

```python
plt.scatter(x_train,y_train, marker='x',color='red')
plt.plot(x_train,pred)
plt.show()
```
