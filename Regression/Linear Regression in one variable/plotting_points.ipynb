```python
#NumPy: Library for scientific computing
#Matplotlib: Library for plotting data
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train is the training example feature values
#y_train is the training example targets 
x_train= np.array([1.0, 2.0])
y_train= np.array([300.0, 500.0])
```

```python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
#m is the number of training examples
m= len(x_train)
print(f"Value of m is: {m}")
```

```python
#so here, to calculate m, we can also use:
print(f"dimensions of x_train : {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
i=1
x_i=x_train[i]
y_i=y_train[i]
print(f"(x_{i},y_{i}) is ({x_i},{y_i})")
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', color='red')
plt.title("Housing Prices")
plt.ylabel('Price')
plt.xlabel('Size')
plt.show()
```

```python
w=100
b=100
```

```python
#np.zeros(n) creates an array of n elements, all initialised with 0

def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=(w*x[i])+b
    return f_wb
```

```python
prediction= compute_model_output(x_train,100,100)
plt.plot(x_train, prediction, color='blue')
plt.show()
```

```python
#the correct values of w and b, that fit our data:  
w=200
b=100
prediction= compute_model_output(x_train,w,b)
plt.plot(x_train, prediction, color='blue', label='Model Prediction')
plt.scatter(x_train, y_train, marker='x', color='red', label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price')
plt.xlabel('Size')
plt.legend()
plt.show()
```

```python
#making predictions using the final model, for:
x_i=1.2

ans= (w*x_i)+b
print(f"For size {x_i}K square feets, the price is ${ans:.0f}k dollars.")
```
